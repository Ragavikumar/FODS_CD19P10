```python
student_data=pd.DataFrame({
    'Name' : ['Alice','Bob','Charlie','David','Eva'],
    'Age' : [23,25,22,24,23],
    'Gender' : ['F','M','M','M','F'],
    'Score' : [85, 78, 92, 70, 88]
})

def grade(Score):
    if Score>=90:
        return 'A'
    elif Score>=80:
        return 'B'
    elif Score>=70:
        return 'C'
    else:
        return 'D'
    
student_data['Grade']=student_data['Score'].apply(grade)
student_data
```

```python
sales_data=pd.DataFrame({
    'Month' : ['Jan','Feb','Mar','Jan','Feb','Mar'],
    'Product' : ['A','A','A','B','B','B'],
    'Sales' : [150, 200, 250, 300, 400, 500]
})
sales_data
```

```python
sales_data=pd.DataFrame({
    'Month' : ['Jan','Feb','Mar','Jan','Feb','Mar'],
    'Product' : ['A','A','A','B','B','B'],
    'Sales' : [150, 200, 250, 300, 400, 500]
})
total_data=sales_data.groupby('Product')['Sales'].sum()
print("Total sales by Product: ")
total_data
```

```python
sales_data=pd.DataFrame({
    'Month' : ['Jan','Feb','Mar','Jan','Feb','Mar'],
    'Product' : ['A','A','A','B','B','B'],
    'Sales' : [150, 200, 250, 300, 400, 500]
})
high=sales_data.loc[sales_data.groupby('Product')['Sales'].idxmax()]
print("Highest sales of each products: ")
high
```

```python
sales_data=pd.DataFrame({
    'Month' : ['Jan','Feb','Mar','Jan','Feb','Mar'],
    'Product' : ['A','A','A','B','B','B'],
    'Sales' : [150, 200, 250, 300, 400, 500]
})
sales_data['Sales_percentage']=sales_data.groupby('Product')['Sales'].apply(lambda x:(x/x.sum())*100)
sales_data
```

```python
# 3. Handling Missing Datas
```

```python
employee_data=pd.DataFrame({
    'Name': ['John','Doe','Jane','Anna','Smith'],
    'Department' : ['HR','Finance',None,'IT','HR'],
    'Salary' : [50000, 60000, 55000, None, 58000]
})
```

```python
employee_data
```

```python
missing=employee_data.isnull().sum()
missing
```

```python
department_mode = employee_data['Department'].mode()[0]
employee_data['Department'].fillna(department_mode, inplace=True)
employee_data
```

```python
salary_mean = employee_data['Salary'].mean()
employee_data['Salary'].fillna(salary_mean, inplace=True)
employee_data
```

```python
employee_data
employee_data.dropna(inplace=True)
employee_data
```
