```python
#The interquartile range (IQR) method to detect and handle outliers in the Salary column.
```

```python
Q1=df['Salary'].quantile(0.25)
```

```python
Q3=df['Salary'].quantile(0.75)
```

```python
IQR=Q3-Q1
```

```python
ib=Q1-1.5*IQR
```

```python
ub=Q3+1.5*IQR
```

```python
outlier=((df[['Salary']]<ib) | (df[['Salary']]>ub)).any(axis=1)
```

```python
cleaned=df[~outlier]
```

```python
cleaned
```
