```python
import pandas as pd
```

```python
df=pd.read_csv('housing_dataset.csv')
```

```python
null_columns = df.columns[df.isnull().any()]
print("Columns with Null Values:")
print(null_columns)
```

```python
median_value=df['Price'].median()
```

```python
df['Price'].fillna(median_value)
```

```python
mean_value=df['Bathrooms'].mean()
```

```python
df['Bathrooms'].fillna(mean_value)
```

```python
df.dropna()
```

```python
# 2.Handling Outliers
```

```python
Q1=df['Price'].quantile(0.25)
Q3=df['Price'].quantile(0.75)
IQR=Q3-Q1
lb=Q1-1.5*IQR
ub=Q3+1.5*IQR
outlier=df[(df['Price']<lb) | (df['Price']>ub)]
outlier
```

```python
df=df[(df['Price']>=lb) & (df['Price']<=ub)]
```

```python
df
```

```python
df['Sqft_Living']=(df['Sqft_Living']-df['Sqft_Living'].min())/(df['Sqft_Living'].max() - df['Sqft_Living'].min())
```

```python
df
```

```python
df['Year_Built']=df['Year_Built'].astype(int)
```

```python
df['Waterfront']=df['Waterfront'].map({'Yes':1,'No':0})
df
```

```python
df.to_csv('cleaned_dataset.csv')
```
